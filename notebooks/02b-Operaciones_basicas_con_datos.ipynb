## Operaciones Básicas con Datos en Python

### 1. Listas (Arrays o Arreglos)

Una **lista** es una colección ordenada de elementos. Piensa en ella como una fila de estantes donde cada estante contiene algo y tiene un número de posición (índice) que empieza en 0.

```python
# Crear una lista de números
precios = [10.50, 20.00, 5.75, 15.20]

# Mostrar la lista
print("Nuestra lista de precios:", precios)
```

#### Acceder a elementos de una lista

Para acceder a un elemento, usamos su índice. Recuerda que el primer elemento está en el índice 0.

```python
# Acceder al primer elemento (índice 0)
primer_precio = precios[0]
print("El primer precio es:", primer_precio)

# Acceder al tercer elemento (índice 2)
tercer_precio = precios[2]
print("El tercer precio es:", tercer_precio)

# Acceder al último elemento (usando un índice negativo)
ultimo_precio = precios[-1]
print("El último precio es:", ultimo_precio)
```

#### Modificar elementos de una lista

Podemos cambiar el valor de un elemento en una posición específica.

```python
# Modificar el segundo precio (índice 1)
precios[1] = 22.00
print("Lista de precios modificada:", precios)
```

#### Añadir elementos a una lista

Usamos el método `.append()` para añadir un elemento al final de la lista.

```python
# Añadir un nuevo precio
precios.append(8.99)
print("Lista de precios con nuevo elemento:", precios)
```

#### Eliminar elementos de una lista

Podemos eliminar elementos por su valor con `.remove()` o por su índice con `del`.

```python
# Eliminar un elemento por su valor (el primero que encuentre)
precios.remove(5.75)
print("Lista después de eliminar 5.75:", precios)

# Eliminar un elemento por su índice (el tercer elemento, que ahora está en el índice 2)
del precios[2]
print("Lista después de eliminar el tercer elemento por índice:", precios)
```

#### Operaciones básicas de análisis con listas de números

Python tiene funciones incorporadas para realizar cálculos sencillos.

```python
numeros = [10, 20, 5, 15, 8, 30]

# Suma de todos los números
suma_total = sum(numeros)
print("Suma total:", suma_total)

# Número más grande (máximo)
maximo = max(numeros)
print("Número máximo:", maximo)

# Número más pequeño (mínimo)
minimo = min(numeros)
print("Número mínimo:", minimo)

# Cantidad de elementos en la lista
cantidad_elementos = len(numeros)
print("Cantidad de elementos:", cantidad_elementos)

# Promedio (media) de los números
promedio = suma_total / cantidad_elementos
print("Promedio:", promedio)
```

### 2. Diccionarios (Maps o Hashmaps)

Un **diccionario** es una colección desordenada de pares `clave: valor`. Piensa en él como una agenda telefónica, donde el nombre de la persona es la clave y su número de teléfono es el valor. Las claves deben ser únicas.

```python
# Crear un diccionario de información de un producto
producto = {
    "nombre": "Laptop",
    "marca": "TechCo",
    "precio": 1200,
    "en_stock": True
}

# Mostrar el diccionario
print("Información del producto:", producto)
```

#### Acceder a valores de un diccionario

Para obtener el valor asociado a una clave, usamos la clave entre corchetes `[]`.

```python
# Acceder al nombre del producto
nombre_producto = producto["nombre"]
print("Nombre del producto:", nombre_producto)

# Acceder al precio
precio_producto = producto["precio"]
print("Precio del producto:", precio_producto)
```

#### Modificar valores de un diccionario

Podemos cambiar el valor asociado a una clave existente.

```python
# Cambiar el precio del producto
producto["precio"] = 1150
print("Diccionario con precio modificado:", producto)
```

#### Añadir nuevos pares clave-valor a un diccionario

Para añadir un nuevo par, simplemente asignamos un valor a una nueva clave.

```python
# Añadir la categoría del producto
producto["categoria"] = "Electrónica"
print("Diccionario con nueva categoría:", producto)
```

#### Eliminar pares clave-valor de un diccionario

Usamos la palabra clave `del` para eliminar un par específico por su clave.

```python
# Eliminar la clave 'en_stock'
del producto["en_stock"]
print("Diccionario después de eliminar 'en_stock':", producto)
```

#### Verificar si una clave existe en un diccionario antes de modificarla

Es importante verificar si una clave ya está presente en el diccionario antes de intentar modificar su valor, para evitar `KeyError` si la clave no existe. También puede ser útil para decidir si añadir una nueva clave o actualizar una existente.

##### Método 1: Usando el operador `in`

El operador `in` devuelve `True` si la clave existe y `False` en caso contrario. Es la forma más idiomática y clara en Python.

```python
# Definimos un diccionario de ejemplo
mi_diccionario = {"nombre": "Ana", "edad": 30, "ciudad": "Madrid"}
print("Diccionario inicial:", mi_diccionario)

# Verificar si la clave 'edad' existe antes de modificarla
if "edad" in mi_diccionario:
    mi_diccionario["edad"] = 31
    print("Clave 'edad' modificada. Nuevo diccionario:", mi_diccionario)
else:
    print("La clave 'edad' no existe en el diccionario.")

# Intentar modificar una clave que no existe
if "pais" in mi_diccionario:
    mi_diccionario["pais"] = "España"
    print("Clave 'pais' modificada. Nuevo diccionario:", mi_diccionario)
else:
    print("La clave 'pais' no existe en el diccionario. Se puede añadir si es necesario.")
    mi_diccionario["pais"] = "España"
    print("Clave 'pais' añadida. Nuevo diccionario:", mi_diccionario)
```

##### Método 2: Usando el método `.get()`

El método `.get(clave, valor_por_defecto)` devuelve el valor de la clave si existe, o un `valor_por_defecto` si no existe (por defecto, `None`). Puedes usar esto para comprobar la existencia y, opcionalmente, asignar un valor predeterminado si la clave no está presente.

```python
# Definimos otro diccionario de ejemplo
animal = {"especie": "Perro", "raza": "Labrador"}
print("Diccionario inicial:", animal)

# Obtener el valor de 'raza'. Si no existe, devolverá None.
raza_actual = animal.get("raza")
if raza_actual is not None:
    print(f"La raza actual es: {raza_actual}")
    animal["raza"] = "Golden Retriever"
    print("Raza modificada. Nuevo diccionario:", animal)
else:
    print("La clave 'raza' no existe.")

# Intentar obtener un valor de una clave que no existe, proporcionando un valor por defecto
color = animal.get("color", "Desconocido")
print(f"El color es: {color}")

# Si queremos añadirla si no existe, podemos combinarlo con el operador 'in'
if "color" not in animal:
    animal["color"] = "Marrón"
    print("Color añadido. Nuevo diccionario:", animal)
```

diferencia entre usar `in` y el método `.get()`
¡Claro! La diferencia principal entre usar el operador in y el método .get() para verificar la existencia de una clave en un diccionario radica en su propósito y lo que devuelven:

Cuándo preferir in:
Solo necesitas saber si la clave existe: Si tu único objetivo es verificar la presencia de una clave en el diccionario (es decir, quieres True o False), el operador in es la forma más directa, legible y pythonica. Es ideal para construir lógica condicional (if 'clave' in diccionario:).

```python
mi_dic = {'a': 1, 'b': 2}
if 'a' in mi_dic:
    print("La clave 'a' existe.") # Solo me importa si existe
```

Cuándo preferir .get():

Necesitas el valor de la clave Y quieres un valor por defecto si no existe: Si además de verificar la existencia, quieres obtener el valor asociado a la clave y proporcionar un valor alternativo si la clave no está presente, .get() es la opción perfecta. Esto te ayuda a evitar KeyError y a manejar casos donde la clave podría faltar de forma elegante.

```python
mi_dic = {'a': 1, 'b': 2}

valor_c = mi_dic.get('c', 0) # Si 'c' no existe, devuelve 0 en lugar de un error
print("Valor de 'c':", valor_c)

valor_a = mi_dic.get('a') # Si 'a' no existe, devuelve None por defecto
print("Valor de 'a':", valor_a)
```

En resumen:

* Usa in cuando tu pregunta es "¿Existe esta clave?" (sí/no).
* Usa .get() cuando tu pregunta es "¿Cuál es el valor de esta clave, y si no está, dame esto otro?".

### Cómo recorrer todas las claves y valores de un diccionario con un bucle

Para recorrer tanto las claves como los valores de un diccionario al mismo tiempo, el método más común y recomendado en Python es usar un bucle `for` junto con el método `.items()` del diccionario.

El método `.items()` devuelve una vista de los pares `(clave, valor)` del diccionario, lo que te permite desempaquetar cada par directamente en dos variables dentro del bucle.

```python
# Usaremos nuestro diccionario 'producto' como ejemplo
# Recordemos su estado actual:
print("Diccionario 'producto' actual:", producto)

print("\n--- Recorriendo el diccionario producto ---")
for clave, valor in producto.items():
    print(f"Clave: {clave}, Valor: {valor}")


# Otro ejemplo con 'mi_diccionario'
print("\nDiccionario 'mi_diccionario' actual:", mi_diccionario)

print("\n--- Recorriendo el diccionario mi_diccionario ---")
for k, v in mi_diccionario.items():
    print(f"La clave '{k}' tiene el valor: {v}")
```

#### Métodos adicionales (menos comunes para ambos a la vez):

- **Solo claves**: Puedes recorrer solo las claves directamente (`for clave in diccionario:`) o usando `.keys()`.
- **Solo valores**: Puedes recorrer solo los valores usando `.values()`.
- **Acceder valores con claves**: Puedes recorrer las claves y luego acceder a los valores usando `diccionario[clave]`, pero `.items()` es más eficiente si necesitas ambos.

```python
# Recorrer solo claves
print("\n--- Solo claves de 'producto' ---")
for clave in producto:
    print(clave)

# Recorrer solo valores
print("\n--- Solo valores de 'producto' ---")
for valor in producto.values():
    print(valor)
```

### Cómo ordenar un diccionario por sus claves o valores

Como mencionamos, los diccionarios en sí mismos no están intrínsecamente 'ordenados' de forma alfabética o numérica, sino por el orden en que se insertaron los elementos. Para obtener una vista ordenada, usamos la función `sorted()` que nos devuelve una lista de los elementos ordenados.

#### 1. Ordenar por Claves

Para ordenar un diccionario por sus claves, puedes usar `sorted()` directamente sobre el diccionario. Por defecto, `sorted()` ordenará las claves en orden ascendente (alfabético para cadenas, numérico para números).

```python
mi_diccionario = {"nombre": "Ana", "edad": 31, "ciudad": "Madrid", "pais": "España"}

print("Diccionario original:", mi_diccionario)

# Obtener las claves ordenadas
claves_ordenadas = sorted(mi_diccionario.keys())
print("\nClaves ordenadas:", claves_ordenadas)

# Recorrer el diccionario por claves ordenadas
print("\n--- Diccionario ordenado por claves ---")
for clave in claves_ordenadas:
    print(f"Clave: {clave}, Valor: {mi_diccionario[clave]}")

# O, de forma más directa, obtener una lista de tuplas (clave, valor) ordenada por clave
diccionario_ordenado_por_clave_items = sorted(mi_diccionario.items())
print("\nLista de tuplas (clave, valor) ordenada por clave:", diccionario_ordenado_por_clave_items)
```

#### 2. Ordenar por Valores

Para ordenar un diccionario por sus valores, necesitas indicarle a `sorted()` qué parte de cada par `(clave, valor)` debe usar para la comparación. Esto se hace con el argumento `key` y una función `lambda`.

```python
diccionario_productos = {
    "Manzanas": 1.20,
    "Plátanos": 0.85,
    "Naranjas": 1.50,
    "Peras": 1.20,
    "Uvas": 2.50
}

print("Diccionario de productos original:", diccionario_productos)

# Ordenar por valores (precio) en orden ascendente
# `lambda item: item[1]` indica que la función de ordenación debe usar el segundo elemento de cada tupla (el valor)
productos_ordenados_por_valor = sorted(diccionario_productos.items(), key=lambda item: item[1])
print("\nProductos ordenados por precio (ascendente):", productos_ordenados_por_valor)

# Ordenar por valores (precio) en orden descendente
productos_ordenados_por_valor_desc = sorted(diccionario_productos.items(), key=lambda item: item[1], reverse=True)
print("\nProductos ordenados por precio (descendente):", productos_ordenados_por_valor_desc)
```

En ambos casos, el resultado es una **lista de tuplas `(clave, valor)`**, no un diccionario nuevo. Si necesitas un diccionario que mantenga este orden (por ejemplo, para Pythons anteriores a 3.7 o para una garantía explícita), podrías reconstruirlo usando `collections.OrderedDict` o simplemente trabajar con la lista de tuplas ordenada.

#### Convertir una lista de tuplas a un diccionario

Como mencionamos, `sorted()` devuelve una lista de tuplas. Si deseas un diccionario que mantenga este nuevo orden (a partir de Python 3.7, los diccionarios conservan el orden de inserción), puedes reconstruirlo fácilmente.

```python
# Tomaremos la lista de tuplas que ordenamos por precio ascendente
print("Lista de tuplas ordenada por precio (ascendente):", productos_ordenados_por_valor)

# Convertir la lista de tuplas de nuevo a un diccionario
diccionario_reconstruido = dict(productos_ordenados_por_valor)

print("\nDiccionario reconstruido a partir de la lista ordenada:", diccionario_reconstruido)

# Podemos verificar que las claves y valores se mantienen y el orden de inserción es el de la lista original.
# Ten en cuenta que, aunque el diccionario conserva el orden de inserción, no es su propósito principal ser una estructura 'ordenada' para operaciones de búsqueda.
```

### Filtrar elementos de un diccionario basados en una condición

Cuando queremos seleccionar solo algunos pares `clave: valor` de un diccionario que cumplen con ciertas reglas, podemos crear un nuevo diccionario. Hay dos maneras principales de hacerlo:

#### 1. Usando una Comprensión de Diccionario (Dictionary Comprehension)

Las comprensiones de diccionario son una forma concisa y elegante de construir nuevos diccionarios a partir de iterables (como otro diccionario o una lista), aplicando transformaciones y/o filtros.

```python
# Usaremos el diccionario de productos como ejemplo
print("Diccionario de productos original:", diccionario_productos)

# Condición: filtrar productos cuyo precio sea mayor o igual a 1.20
productos_caros = {clave: valor for clave, valor in diccionario_productos.items() if valor >= 1.20}

print("\nProductos con precio >= 1.20:", productos_caros)

# Otro ejemplo: filtrar productos por nombre que contenga la letra 'a'
productos_con_a = {clave: valor for clave, valor in diccionario_productos.items() if 'a' in clave.lower()}

print("\nProductos cuyo nombre contiene 'a':", productos_con_a)
```

#### 2. Usando un bucle `for` y una sentencia `if`

Esta es una forma más explícita de lograr el mismo resultado, especialmente útil si la lógica de filtrado es más compleja o si estás empezando con las comprensiones.

```python
# Volvemos a usar el diccionario de productos
print("Diccionario de productos original:", diccionario_productos)

productos_baratos = {}
for clave, valor in diccionario_productos.items():
    if valor < 1.20:
        productos_baratos[clave] = valor

print("\nProductos con precio < 1.20 (usando bucle for):", productos_baratos)

# Otro ejemplo: filtrar por clave (nombre de producto) que no empiece por 'M'
productos_no_m = {}
for clave, valor in diccionario_productos.items():
    if not clave.startswith('M'):
        productos_no_m[clave] = valor

print("\nProductos que no empiezan por 'M':", productos_no_m)
```

### Filtrar elementos de un diccionario basados en múltiples condiciones

Para aplicar varias condiciones al filtrar un diccionario, usamos los operadores lógicos `and` (y), `or` (o) y `not` (no). Estos operadores nos permiten construir condiciones más complejas. Seguimos aplicando el filtro para crear un nuevo diccionario.

#### 1. Usando Comprensión de Diccionario con múltiples condiciones

```python
# Usaremos el diccionario de productos como ejemplo
print("Diccionario de productos original:", diccionario_productos)

# Condición 1: precio mayor o igual a 1.20
# Condición 2: el nombre del producto NO contiene la letra 'a' (ignorando mayúsculas/minúsculas)
productos_filtrados_and = {
    clave: valor for clave, valor in diccionario_productos.items()
    if valor >= 1.20 and 'a' not in clave.lower()
}
print("\nProductos con precio >= 1.20 Y sin 'a' en el nombre:", productos_filtrados_and)

# Condición 1: precio menor que 1.00
# Condición 2: el nombre del producto empieza por 'P'
productos_filtrados_or = {
    clave: valor for clave, valor in diccionario_productos.items()
    if valor < 1.00 or clave.startswith('P')
}
print("\nProductos con precio < 1.00 O que empiezan por 'P':", productos_filtrados_or)
```

#### 2. Usando un bucle `for` y una sentencia `if` con múltiples condiciones

```python
# Volvemos a usar el diccionario de productos
print("Diccionario de productos original:", diccionario_productos)

# Queremos productos que valgan entre 1.00 y 2.00 (ambos incluidos)
productos_rango_precio = {}
for clave, valor in diccionario_productos.items():
    if valor >= 1.00 and valor <= 2.00:
        productos_rango_precio[clave] = valor

print("\nProductos con precio entre 1.00 y 2.00 (bucle for):", productos_rango_precio)

# Otro ejemplo: productos que no sean 'Manzanas' Y que su precio sea menor a 1.50
productos_especificos = {}
for clave, valor in diccionario_productos.items():
    if clave != 'Manzanas' and valor < 1.50:
        productos_especificos[clave] = valor

print("\nProductos que no son 'Manzanas' y precio < 1.50 (bucle for):", productos_especificos)
```

### Diccionarios Anidados: Diccionarios dentro de otros Diccionarios

Un **diccionario anidado** es un diccionario donde el valor de una o más de sus claves es, a su vez, otro diccionario. Esto nos permite representar estructuras de datos más complejas, como registros con subcategorías o bases de datos simples.

```python
# Ejemplo de un diccionario anidado: Información de diferentes usuarios
usuarios = {
    "ana_perez": {
        "nombre": "Ana",
        "apellido": "Pérez",
        "edad": 28,
        "ciudad": "Madrid",
        "intereses": ["lectura", "senderismo", "cocina"]
    },
    "juan_garcia": {
        "nombre": "Juan",
        "apellido": "García",
        "edad": 35,
        "ciudad": "Barcelona",
        "intereses": ["programación", "cine", "viajes"]
    },
    "maria_lopez": {
        "nombre": "María",
        "apellido": "López",
        "edad": 22,
        "ciudad": "Valencia",
        "intereses": ["pintura", "música", "deportes"]
    }
}

print("Diccionario de usuarios anidado:")
# Para una mejor visualización, puedes usar json.dumps
import json
print(json.dumps(usuarios, indent=4))
```

#### Acceder a elementos en diccionarios anidados

Para acceder a un valor dentro de un diccionario anidado, simplemente encadenas las claves, una tras otra, como si estuvieras 'navegando' por la estructura.

```python
# Acceder al nombre de Ana
nombre_ana = usuarios["ana_perez"]["nombre"]
print(f"Nombre de Ana: {nombre_ana}")

# Acceder a la ciudad de Juan
ciudad_juan = usuarios["juan_garcia"]["ciudad"]
print(f"Ciudad de Juan: {ciudad_juan}")

# Acceder al primer interés de María
primer_interes_maria = usuarios["maria_lopez"]["intereses"][0]
print(f"Primer interés de María: {primer_interes_maria}")

# Modificar la edad de Ana
usuarios["ana_perez"]["edad"] = 29
print(f"Nueva edad de Ana: {usuarios['ana_perez']['edad']}")

# Añadir un nuevo interés a Juan
usuarios["juan_garcia"]["intereses"].append("ajedrez")
print(f"Nuevos intereses de Juan: {usuarios['juan_garcia']['intereses']}")
```

### Eliminar una clave de un diccionario anidado

Para eliminar una clave de un diccionario anidado, primero debes acceder al diccionario interno que contiene la clave que quieres borrar. Una vez que tengas una referencia a ese diccionario interno, puedes usar la palabra clave `del` de la misma manera que lo harías con un diccionario simple.

```python
# Recordemos la estructura actual de nuestro diccionario 'usuarios'
print("Diccionario 'usuarios' antes de la eliminación:")
import json
print(json.dumps(usuarios, indent=4))

# Queremos eliminar la clave 'ciudad' del usuario 'maria_lopez'
# Primero, accedemos al diccionario de María
usuario_maria = usuarios['maria_lopez']

# Ahora, eliminamos la clave 'ciudad' de su diccionario
del usuario_maria['ciudad']

print("\nDiccionario 'usuarios' después de eliminar la clave 'ciudad' de María:")
print(json.dumps(usuarios, indent=4))

# También puedes hacerlo directamente, sin una variable intermedia:
del usuarios['juan_garcia']['intereses'] # Eliminar la lista de intereses de Juan

print("\nDiccionario 'usuarios' después de eliminar la clave 'intereses' de Juan:")
print(json.dumps(usuarios, indent=4))
```

### Añadir una nueva clave a un usuario en el diccionario anidado

Para añadir una nueva clave (por ejemplo, 'email') a un usuario existente, como 'ana_perez', simplemente accedes a su diccionario y le asignas un valor a la nueva clave. Python creará automáticamente la clave si no existe.

```python
# Recordemos el estado actual del diccionario 'usuarios'
print("Diccionario 'usuarios' antes de añadir la nueva clave:")
import json
print(json.dumps(usuarios, indent=4))

# Añadir una clave 'email' al usuario 'ana_perez'
usuarios['ana_perez']['email'] = 'ana.perez@example.com'

# Añadir una clave 'profesion' al usuario 'maria_lopez'
usuarios['maria_lopez']['profesion'] = 'Diseñadora gráfica'

print("\nDiccionario 'usuarios' después de añadir nuevas claves:")
print(json.dumps(usuarios, indent=4))
```

### Convertir un diccionario anidado a un DataFrame de Pandas

Para convertir un diccionario anidado donde las claves de nivel superior representan entidades (en este caso, usuarios) y los valores son diccionarios con atributos de esas entidades, `pandas.DataFrame.from_dict()` con el argumento `orient='index'` es la forma más eficiente. Esto creará un DataFrame donde cada usuario es una fila y sus atributos son columnas.

```python
import pandas as pd

# El diccionario 'usuarios' ya está definido y modificado de pasos anteriores

# Convertir el diccionario anidado a un DataFrame
# 'orient="index"' indica que las claves de nivel superior deben ser el índice del DataFrame
df_usuarios = pd.DataFrame.from_dict(usuarios, orient='index')

# Mostrar el DataFrame resultante
print("DataFrame de Usuarios:")
display(df_usuarios)

# Para ver la información general del DataFrame (tipos de datos, valores no nulos)
print("\nInformación del DataFrame:")
df_usuarios.info()
```
