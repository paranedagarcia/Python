### Similitudes y Diferencias Clave

Las tuplas y las listas son ambas colecciones ordenadas de elementos en Python. Sin embargo, tienen una diferencia fundamental que las distingue:

#### Listas (`list`)

*   **Mutables**: Esto significa que puedes cambiar, añadir o eliminar elementos después de que la lista ha sido creada.
*   Se definen utilizando **corchetes `[]`**.
*   Son ideales para colecciones de elementos que necesitan ser modificadas con frecuencia.


````python
# Ejemplo de Lista (Mutable)
mi_lista = [1, 2, 3, 'manzana', True]
print(f"Lista original: {mi_lista}")

# Podemos modificar elementos
mi_lista[0] = 100
print(f"Lista después de cambiar un elemento: {mi_lista}")

# Podemos añadir elementos
mi_lista.append('naranja')
print(f"Lista después de añadir un elemento: {mi_lista}")

# Podemos eliminar elementos
mi_lista.remove(True)
print(f"Lista después de eliminar un elemento: {mi_lista}")
````

#### Tuplas (`tuple`)

*   **Inmutables**: Esto significa que, una vez que una tupla ha sido creada, no puedes cambiar, añadir o eliminar sus elementos. Si necesitas una versión modificada, debes crear una tupla nueva.
*   Se definen utilizando **paréntesis `()`** (aunque los paréntesis son opcionales si hay al menos dos elementos).
*   Son ideales para colecciones de elementos que no deben cambiar, como coordenadas geográficas, fechas, o cuando una función devuelve un conjunto fijo de resultados.


````python
# Ejemplo de Tupla (Inmutable)
mi_tupla = (1, 2, 3, 'pera', False)
print(f"Tupla original: {mi_tupla}")

# Intentar modificar un elemento causará un error
try:
    mi_tupla[0] = 100
except TypeError as e:
    print(f"Error al intentar modificar un elemento de la tupla: {e}")

# Intentar añadir un elemento también causará un error
try:
    mi_tupla.append('limon')
except AttributeError as e:
    print(f"Error al intentar añadir un elemento a la tupla: {e}")

# Para 'modificar' una tupla, debes crear una nueva
mi_nueva_tupla = mi_tupla + (4,)
print(f"Nueva tupla creada a partir de la anterior: {mi_nueva_tupla}")
````

#### Resumen de Diferencias:

| Característica | Lista (`[]`)       | Tupla (`()`)       |
|----------------|--------------------|--------------------|
| **Mutabilidad**| Mutable (se puede cambiar) | Inmutable (no se puede cambiar) |
| **Sintaxis**   | Corchetes `[]`     | Paréntesis `()`    |
| **Uso Típico** | Colecciones que cambian | Colecciones fijas, registro de datos |
| **Rendimiento**| Ligeramente más lenta | Ligeramente más rápida (algunas operaciones) |

En resumen, la elección entre una lista y una tupla depende de si necesitas que la colección de elementos sea modificable o no después de su creación. Si los datos son fijos, una tupla es una buena elección por su inmutabilidad y eficiencia. Si los datos van a cambiar, una lista es la opción correcta.

### 6. Conversión entre Listas y Tuplas

Python facilita la conversión entre estas dos estructuras de datos usando las funciones `list()` y `tuple()`.


````python
# Convertir una tupla en una lista
mi_tupla_original = (10, 20, 30, 'texto')
lista_desde_tupla = list(mi_tupla_original)

print(f"Tupla original: {mi_tupla_original} (Tipo: {type(mi_tupla_original)})")
print(f"Lista convertida: {lista_desde_tupla} (Tipo: {type(lista_desde_tupla)})")

# Ahora que es una lista, podemos modificarla
lista_desde_tupla.append(40)
print(f"Lista modificada: {lista_desde_tupla}")
````

**Cuándo es mejor usar una lista en lugar de una tupla**

La elección entre una lista y una tupla en Python se reduce principalmente a la mutabilidad y al propósito de la colección de datos. Cuándo es mejor usar una lista:

**Cuándo es Mejor Usar una Lista ([])**:

**Necesitas una colección modificable**: Este es el caso de uso más importante. Si los datos de tu colección van a cambiar después de ser creados (añadir, eliminar, modificar elementos), una lista es la elección correcta. Ejemplos incluyen:

- Un registro de compras: A medida que añades o quitas productos del carrito.
- Una lista de tareas pendientes: Cuando añades nuevas tareas o marcas algunas como completadas y las eliminas.
- Resultados intermedios de un cálculo: Donde vas acumulando valores que luego vas a procesar o filtrar.

- El orden de los elementos es importante y puede variar: Aunque ambas son ordenadas, las listas son flexibles para reordenar o insertar elementos en cualquier posición.

- Tienes una colección homogénea de ítems: A menudo, las listas se usan para almacenar muchos elementos del mismo tipo (aunque Python permite mezclar tipos), como una lista de nombres de usuarios, una lista de números, etc.

- Estás usando la colección para implementar estructuras de datos dinámicas: Como pilas (stacks), colas (queues) o cualquier estructura donde los elementos se añaden y eliminan con frecuencia.

- En resumen: Usa listas cuando preveas que la colección de datos va a cambiar o evolucionar a lo largo del tiempo de vida de tu programa.

**Cuándo es Mejor Usar una Tupla (())**:

- Necesitas una colección inmutable: Si la colección de elementos no debe cambiar una vez creada (por seguridad, integridad de datos o simplemente porque sabes que los datos son fijos). Ejemplos incluyen:

-  Coordenadas geográficas (latitud, longitud): Un punto en el mapa es fijo.
Un registro de base de datos: Donde cada fila es un conjunto de datos relacionados que no debe alterarse.
- Los resultados de una función que devuelve múltiples valores: Como vimos antes, Python los empaqueta automáticamente en una tupla, porque son un conjunto de resultados fijos para una operación.
- La colección es de tamaño fijo y el orden de los elementos es importante pero no cambiará: Los elementos tienen un significado posicional específico y ese significado no se alterará.

- Tienes una colección heterogénea de ítems con significado posicional: Es decir, los elementos pueden ser de diferentes tipos, pero cada posición dentro de la tupla tiene un significado particular (ej: (nombre, edad, email)).

- Quieres usar la colección como clave en un diccionario o como elemento en un conjunto: Las tuplas, al ser inmutables, son 'hashable' y pueden usarse en estos contextos, a diferencia de las listas.

- Pequeñas optimizaciones de rendimiento: En algunos casos, las tuplas pueden ser marginalmente más rápidas que las listas para operaciones de acceso, y consumen un poco menos de memoria, aunque esto raramente es un factor decisivo en la mayoría de las aplicaciones.

En resumen: Usa tuplas cuando la colección de datos es fija y no debe ser modificada después de su creación.

### Tupla 'hashable'

Entender qué significa que una tupla sea 'hashable' es clave para comprender sus usos específicos, especialmente en comparación con las listas. Es una de las propiedades más importantes de la inmutabilidad.

En Python, un objeto es '**hashable**' si tiene un **valor hash** que nunca cambia durante su vida útil (es decir, es inmutable) y puede compararse con otros objetos.

- **Valor Hash**: Es un número entero que identifica de forma única a un objeto durante su tiempo de vida. Piensa en ello como una huella digital o un código de barras de un objeto. Python usa este valor hash internamente para almacenar y recuperar objetos de manera muy eficiente en estructuras de datos específicas.

- **Inmutabilidad**: Para que un objeto sea hashable, debe ser **inmutable**. Esto significa que su contenido no puede cambiar una vez que ha sido creado. Si un objeto cambiara, su valor hash también debería cambiar, lo cual rompería la lógica de las estructuras de datos que dependen de la hashabilidad.

  - Tuplas son hashable (si todos sus elementos también lo son) porque son inmutables.
  - Listas no son hashable porque son mutables. Su contenido puede cambiar, lo que significaría que su valor hash podría cambiar, invalidando su uso en ciertos contextos.

#### La Hashabilidad

Los objetos hashable son fundamentales para dos tipos de estructuras de datos en Python:

- Claves de Diccionarios (dict): Las claves de un diccionario deben ser objetos hashable. Cuando creas un diccionario, Python usa el valor hash de las claves para almacenarlas y luego recuperarlas rápidamente. Si una clave no fuera hashable (como una lista), no podría usarse de manera eficiente como clave.


````python
# Esto es válido: la tupla es hashable
mi_diccionario = {
    (1, 2): 'Coordenada A',
    (3, 4): 'Coordenada B'
}
print(mi_diccionario[(1, 2)])

# Esto NO es válido: la lista no es hashable
# otro_diccionario = {
#     [1, 2]: 'Valor'
# }
# TypeError: unhashable type: 'list'
````

- Elementos de Conjuntos (set): Los conjuntos son colecciones de elementos únicos y no ordenados. Para garantizar la unicidad y para su funcionamiento interno, todos los elementos de un conjunto deben ser objetos hashable.


````python
# Esto es válido: la tupla es hashable
mi_conjunto = {
    (1, 2),
    (3, 4),
    (1, 2) # El duplicado se ignora
}
print(mi_conjunto) # Salida: {(3, 4), (1, 2)}

# Esto NO es válido: la lista no es hashable
# otro_conjunto = {
#     [1, 2],
#     [3, 4]
# }
# TypeError: unhashable type: 'list'
````

En Resumen:

Que una tupla sea 'hashable' significa que es inmutable y, por lo tanto, Python puede calcular un valor hash único para ella. Esta propiedad la hace adecuada para ser utilizada como claves en diccionarios o como elementos en conjuntos, lo cual no es posible con las listas debido a su mutabilidad.

### Ejemplo de Tuplas como Elementos de un Conjunto (`set`)

Como mencionamos, la inmutabilidad de las tuplas les permite ser usadas como elementos de conjuntos o claves de diccionarios. Esto no sería posible con las listas.


````python
# Definimos algunas tuplas que representarán, por ejemplo, puntos en un plano (x, y)
punto1 = (1, 2)
punto2 = (3, 4)
punto3 = (1, 2) # Este es un duplicado del punto1
punto4 = (5, 6)

# Creamos un conjunto de puntos
conjunto_de_puntos = {punto1, punto2, punto3, punto4}

print(f"Conjunto de puntos: {conjunto_de_puntos}")
print(f"Tipo del conjunto: {type(conjunto_de_puntos)}")

# Como los conjuntos solo almacenan elementos únicos, el punto3 (que es igual a punto1)
# no se añade dos veces. Esto demuestra la unicidad y el uso de hash para comparar.
print(f"Número de elementos en el conjunto: {len(conjunto_de_puntos)}")

print("\n---\n")

# Intentar añadir una lista a un conjunto causaría un error
lista_punto = [7, 8]
try:
    conjunto_con_lista = {punto1, lista_punto}
except TypeError as e:
    print(f"Error al intentar añadir una lista a un conjunto: {e}")
````

Este ejemplo ilustra cómo las tuplas, al ser inmutables y, por ende, hashable, encajan perfectamente en la definición de lo que un conjunto requiere para sus elementos. La lista, al ser mutable, no puede ofrecer esa garantía de estabilidad de hash, y por eso Python nos impide añadirla directamente a un conjunto.
