## Operaciones con Cadenas en Python

Las cadenas (strings) son secuencias de caracteres y son un tipo de dato fundamental en Python. Permiten almacenar texto y realizar diversas operaciones con él. A continuación, se presentan varios ejemplos de operaciones comunes con cadenas.

### 1. Concatenación de Cadenas

```python
# Usando el operador '+'
cadena1 = "Hola"
cadena2 = "Mundo"
cadena_concatenada = cadena1 + " " + cadena2
print(f"Concatenación con '+': {cadena_concatenada}")

# Usando f-strings (una forma moderna y recomendada)
nombre = "Alice"
edad = 30
mensaje_fstring = f"Mi nombre es {nombre} y tengo {edad} años."
print(f"Concatenación con f-string: {mensaje_fstring}")
```

#### Explicación:

*   El operador `+` se utiliza para unir dos o más cadenas, creando una nueva cadena a partir de ellas.
*   Las f-strings (formatted string literals) son una manera eficiente y legible de incrustar expresiones de Python dentro de cadenas literales. Se prefieren para la mayoría de los casos de formateo de cadenas.

### 2. Longitud de una Cadena

```python
texto = "Python es divertido si se aplica bien!"
longitud = len(texto)
print(f"La longitud de la cadena es: {longitud}")
```

#### Explicación:

*   La función `len()` devuelve el número de caracteres en una cadena, incluyendo espacios y caracteres especiales.

### 3. Acceso a Caracteres (Indexación)

```python
frase = "Programación basada en procedimientos"

# Acceso al primer carácter (índice 0)
primer_caracter = frase[0]
print(f"Primer carácter: {primer_caracter}")

# Acceso al último carácter (índice -1)
ultimo_caracter = frase[-1]
print(f"Último carácter: {ultimo_caracter}")

# Acceso a un carácter específico (por ejemplo, el cuarto carácter)
cuarto_caracter = frase[3]
print(f"Cuarto carácter: {cuarto_caracter}")
```

#### Explicación:

*   Las cadenas son secuencias indexadas, lo que significa que puedes acceder a caracteres individuales utilizando su posición (índice).
*   La indexación comienza en `0` para el primer carácter.
*   La indexación negativa permite acceder a caracteres desde el final de la cadena, donde `-1` es el último carácter, `-2` es el penúltimo, y así sucesivamente.

### 4. Rebanado de Cadenas (Slicing)

```python
frase = "Manipulación de datos"

# Obtener una subcadena desde el índice 0 hasta (sin incluir) el 4
subcadena1 = frase[0:4]
print(f"Slice [0:4]: {subcadena1}")

# Obtener una subcadena desde el índice 5 hasta el final
subcadena2 = frase[5:]
print(f"Slice [5:]: {subcadena2}")

# Obtener una subcadena desde el principio hasta (sin incluir) el 6
subcadena3 = frase[:6]
print(f"Slice [:6]: {subcadena3}")

# Obtener una subcadena cada 2 caracteres
subcadena_paso = frase[::2]
print(f"Slice [::2]: {subcadena_paso}")

# Invertir la cadena
cadena_invertida = frase[::-1]
print(f"Slice invertido [::-1]: {cadena_invertida}")
```

#### Explicación:

*   El rebanado (slicing) permite extraer subcadenas utilizando la sintaxis `[inicio:fin:paso]`.
    *   `inicio`: El índice donde comienza la rebanada (incluido). Si se omite, se asume el principio de la cadena.
    *   `fin`: El índice donde termina la rebanada (excluido). Si se omite, se asume el final de la cadena.
    *   `paso`: El número de caracteres a saltar. Si se omite, se asume `1`.
*   Un `paso` negativo (como `-1`) invierte la cadena.

### 5. Métodos de Cadenas: Cambio de Mayúsculas/Minúsculas

```python
texto_original = "pYThoN PrOgRAmMing tools"

# Todo a mayúsculas
mayusculas = texto_original.upper()
print(f"Mayúsculas: {mayusculas}")

# Todo a minúsculas
minusculas = texto_original.lower()
print(f"Minúsculas: {minusculas}")

# Primera letra de la cadena en mayúscula
capitalizada = texto_original.capitalize()
print(f"Capitalizada: {capitalizada}")

# Primera letra de cada palabra en mayúscula
titulo = texto_original.title()
print(f"Formato título: {titulo}")
```

#### Explicación:

*   `upper()`: Convierte todos los caracteres de la cadena a mayúsculas.
*   `lower()`: Convierte todos los caracteres de la cadena a minúsculas.
*   `capitalize()`: Convierte el primer carácter de la cadena a mayúscula y el resto a minúsculas.
*   `title()`: Convierte la primera letra de cada palabra de la cadena a mayúscula y el resto a minúsculas.

### 6. Métodos de Cadenas: Búsqueda y Reemplazo

```python
oracion = "El perro es un buen amigo, y el gato también."

# Buscar la primera ocurrencia de una subcadena
indice_perro = oracion.find("perro")
print(f"'perro' encontrado en la oración: {indice_perro}")

# Si no se encuentra, find() devuelve -1
indice_pez = oracion.find("pez")
print(f"'pez' encontrado en la oración: {indice_pez}")

# Reemplazar subcadenas
oracion_modificada = oracion.replace("perro", "conejo")
print(f"Reemplazando 'perro' por 'conejo': {oracion_modificada}")

# Reemplazar solo las primeras N ocurrencias
oracion_limit_replace = oracion.replace("el", "un", 1)
print(f"Reemplazando solo la primera 'el': {oracion_limit_replace}")
```

#### Explicación:

*   `find(subcadena)`: Devuelve el índice de la primera ocurrencia de `subcadena` dentro de la cadena. Si no la encuentra, devuelve `-1`.
*   `replace(viejo, nuevo, [cuenta])`: Reemplaza todas las ocurrencias de `viejo` por `nuevo`. Opcionalmente, `cuenta` especifica el número máximo de ocurrencias a reemplazar.

### 7. Métodos de Cadenas: Dividir y Unir

```python
datos = "manzana,pera,uva,platano"

# Dividir una cadena en una lista usando un delimitador
lista_frutas = datos.split(',')
print(f"Lista de frutas (split por ','): {lista_frutas}")

frase_espacios = "Esto es una frase con espacios"
lista_palabras = frase_espacios.split()
print(f"Lista de palabras (split por espacio por defecto): {lista_palabras}")

# Unir elementos de una lista en una cadena usando un separador
separador = " - "
cadena_unida = separador.join(lista_frutas)
print(f"Cadena unida con ' - ': {cadena_unida}")
```

#### Explicación:

*   `split(delimitador)`: Divide la cadena en una lista de subcadenas utilizando el `delimitador` especificado. Si no se proporciona un delimitador, divide por espacios en blanco.
*   `separador.join(iterable)`: Une los elementos de un iterable (como una lista) en una sola cadena, insertando el `separador` entre cada elemento.

### 8. Métodos de Cadenas: Verificación de Contenido

```python
s1 = "12345"
s2 = "HolaMundo"
s3 = "Python3"
s4 = "   "
s5 = "Hola Mundo"

print(f"'{s1}'.isdigit(): {s1.isdigit()}")
print(f"'{s2}'.isalpha(): {s2.isalpha()}")
print(f"'{s3}'.isalnum(): {s3.isalnum()}")
print(f"'{s4}'.isspace(): {s4.isspace()}")
print(f"'{s5}'.isspace(): {s5.isspace()}") # Contiene espacio, por lo tanto False
print(f"'{s5}'.startswith('Hola'): {s5.startswith('Hola')}")
print(f"'{s5}'.endswith('Mundo'): {s5.endswith('Mundo')}")
```

#### Explicación:

Estos métodos devuelven `True` o `False` dependiendo del contenido de la cadena:

*   `isdigit()`: `True` si todos los caracteres son dígitos y hay al menos un carácter, `False` en caso contrario.
*   `isalpha()`: `True` si todos los caracteres son letras alfabéticas y hay al menos un carácter, `False` en caso contrario.
*   `isalnum()`: `True` si todos los caracteres son alfanuméricos (letras o números) y hay al menos un carácter, `False` en caso contrario.
*   `isspace()`: `True` si todos los caracteres son espacios en blanco y hay al menos un carácter, `False` en caso contrario.
*   `startswith(prefijo)`: `True` si la cadena comienza con `prefijo`.
*   `endswith(sufijo)`: `True` si la cadena termina con `sufijo`.

### 9. Eliminación de Espacios en Blanco (Strip)

```python
cadena_con_espacios = "   Hola, mundo!   "

# Eliminar espacios en blanco al principio y al final
cadena_limpia = cadena_con_espacios.strip()
print(f"Cadena limpia: '{cadena_limpia}'")

# Eliminar solo espacios a la izquierda
cadena_lstrip = cadena_con_espacios.lstrip()
print(f"Cadena lstrip: '{cadena_lstrip}'")

# Eliminar solo espacios a la derecha
cadena_rstrip = cadena_con_espacios.rstrip()
print(f"Cadena rstrip: '{cadena_rstrip}'")
```

#### Explicación:

*   `strip()`: Elimina cualquier espacio en blanco (o caracteres especificados) del principio y del final de la cadena.
*   `lstrip()`: Elimina los espacios en blanco (o caracteres especificados) solo del lado izquierdo de la cadena.
*   `rstrip()`: Elimina los espacios en blanco (o caracteres especificados) solo del lado derecho de la cadena.

### 10. Filtrar una Lista de Cadenas Usando `startswith()` y `strip()`

```python
lista_cadenas = [
    "  Apple",
    "banana  ",
    "  Apricot ",
    "Grape",
    "avocado",
    "APple Pie"
]

prefijo_buscado = "ap"

# Primero, limpiar los espacios en blanco de cada cadena y luego filtrar (sin distinción de mayúsculas/minúsculas)
cadenas_filtradas = []
for cadena in lista_cadenas:
    cadena_limpia = cadena.strip()  # Eliminar espacios en blanco alrededor
    if cadena_limpia.lower().startswith(prefijo_buscado.lower()): # Convertir a minúsculas antes de comparar
        cadenas_filtradas.append(cadena_limpia)

print(f"Lista original: {lista_cadenas}")
print(f"Prefijo buscado (sin distinguir mayúsculas/minúsculas): '{prefijo_buscado}'")
print(f"Cadenas que empiezan con '{prefijo_buscado}' (después de limpiar y sin distinguir mayúsculas/minúsculas): {cadenas_filtradas}")

# Otra forma más concisa usando comprensión de listas
cadenas_filtradas_comprehension = [
    cadena.strip() for cadena in lista_cadenas
    if cadena.strip().lower().startswith(prefijo_buscado.lower())
]
print(f"Cadenas filtradas con comprensión de listas (sin distinguir mayúsculas/minúsculas): {cadenas_filtradas_comprehension}")
```

#### Explicación:

Este ejemplo demuestra cómo combinar `strip()` y `startswith()` para procesar una lista de cadenas, ahora **sin distinguir entre mayúsculas y minúsculas**:

1.  **`lista_cadenas`**: Se define una lista de cadenas, algunas de las cuales tienen espacios en blanco al principio o al final, y diferentes combinaciones de mayúsculas y minúsculas.
2.  **`prefijo_buscado`**: Se define el prefijo que queremos buscar.
3.  **Bucle `for`**: Se itera sobre cada `cadena` en `lista_cadenas`.
    *   `cadena.strip()`: Antes de realizar la verificación, se usa `strip()` para eliminar cualquier espacio en blanco no deseado de los extremos de la cadena. Esto asegura que la comparación de `startswith()` sea precisa y no se vea afectada por los espacios.
    *   `cadena_limpia.lower().startswith(prefijo_buscado.lower())`: Aquí es donde se añade la lógica de insensibilidad a mayúsculas y minúsculas. Tanto `cadena_limpia` como `prefijo_buscado` se convierten a minúsculas utilizando el método `.lower()` antes de realizar la comparación con `startswith()`. De esta forma, 'Apple', 'apple', y 'APple' serán tratados de la misma manera que 'ap'.
    *   Si la condición es `True`, la cadena limpia se añade a `cadenas_filtradas`.
4.  **Comprensión de listas**: Se muestra una forma más concisa y 'pythónica' de lograr el mismo resultado utilizando una comprensión de listas, aplicando `strip()`, `lower()` y `startswith()` en una sola línea.

otros métodos existen en Python para comparar cadenas de texto


1. Comparación Directa de Igualdad (==, !=)

- == (igual a): Comprueba si dos cadenas son exactamente iguales (sensible a mayúsculas/minúsculas y espacios).
- != (distinto de): Comprueba si dos cadenas son diferentes.

Ejemplo:


```python
cadena_a = "Hola"
cadena_b = "hola"
cadena_c = "Hola"

print(f"'cadena_a' == 'cadena_b': {cadena_a == cadena_b}") # False
print(f"'cadena_a' == 'cadena_c': {cadena_a == cadena_c}") # True
print(f"'cadena_a' != 'cadena_b': {cadena_a != cadena_b}") # True
```

2. Comparación Lexicográfica (<, >, <=, >=)

Estas operaciones comparan cadenas basándose en el orden alfabético (ASCII/Unicode). Se compara carácter a carácter.

Ejemplo:

```python
print(f"'apple' < 'banana': {'apple' < 'banana'}") # True
print(f"'Zebra' < 'apple': {'Zebra' < 'apple'}")   # True (Z tiene un valor ASCII menor que a)
print(f"'10' > '2': {'10' > '2'}")                 # False (compara el primer carácter '1' con '2')
```

3. El Operador in (Contiene Subcadena)

Este operador verifica si una subcadena está presente dentro de otra cadena. Devuelve True si la encuentra, False en caso contrario.

Ejemplo:

```python
frase = "El rápido zorro marrón saltó sobre el perro perezoso."

print(f"'zorro' in frase: {'zorro' in frase}") # True
print(f"'gato' in frase: {'gato' in frase}")   # False

# Es sensible a mayúsculas/minúsculas por defecto
print(f"'Zorro' in frase: {'Zorro' in frase}") # False
# Para hacerla insensible a mayúsculas/minúsculas:
print(f"'Zorro'.lower() in frase.lower(): {'Zorro'.lower() in frase.lower()}") # True
```

4. endswith() (Termina con)

Similar a startswith(), pero verifica si una cadena termina con un sufijo específico.

Ejemplo:

```python
nombre_archivo = "documento.pdf"
print(f"'{nombre_archivo}'.endswith('.pdf'): {nombre_archivo.endswith('.pdf')}") # True
print(f"'{nombre_archivo}'.endswith('.txt'): {nombre_archivo.endswith('.txt')}") # False
```

5. Comparaciones Insensibles a Mayúsculas/Minúsculas

Como ya hemos visto con startswith(), para realizar comparaciones sin distinguir mayúsculas de minúsculas, la estrategia común es convertir ambas cadenas a un formato común (todo minúsculas o todo mayúsculas) antes de la comparación.

Ejemplo (recapitulación):

```python
cadena1 = "Python"
cadena2 = "python"

print(f"cadena1.lower() == cadena2.lower(): {cadena1.lower() == cadena2.lower()}") # True
```

6. Expresiones Regulares (re Module)

Para patrones de búsqueda y comparación más complejos, el módulo re (regular expressions) es extremadamente potente. Permite definir patrones sofisticados.

Ejemplo:

```python
import re

texto = "Mi número de teléfono es 123-456-7890."

# Buscar un patrón de número de teléfono
patron = r"\d{3}-\d{3}-\d{4}"

if re.search(patron, texto):
    print(f"Patrón de teléfono encontrado en: '{texto}'")
else:
    print(f"Patrón de teléfono no encontrado en: '{texto}'")

# Búsqueda insensible a mayúsculas/minúsculas con re.IGNORECASE
if re.search("hola", "HOLA", re.IGNORECASE):
    print("'HOLA' coincide con 'hola' (ignorando mayúsculas/minúsculas)")
```

7. Comparaciones Locales (locale Module)

Para aplicaciones que necesitan comparar cadenas según las reglas culturales y lingüísticas de un idioma (por ejemplo, para ordenar palabras con acentos o caracteres especiales de manera correcta en español, francés, etc.), Python ofrece el módulo locale.

Ejemplo (requiere configurar el locale):

```python
import locale

# Intentar configurar el locale (puede variar según el sistema operativo)
try:
    locale.setlocale(locale.LC_ALL, 'es_ES.UTF-8')
except locale.Error:
    print("Advertencia: No se pudo configurar el locale 'es_ES.UTF-8'. Las comparaciones sensibles al locale pueden no funcionar como se espera.")
    # Fallback para sistemas que no tienen 'es_ES.UTF-8'
    try:
        locale.setlocale(locale.LC_ALL, 'Spanish_Spain.1252')
    except locale.Error:
        print("Advertencia: No se pudo configurar el locale 'Spanish_Spain.1252'.")

cadena1 = "águila"
cadena2 = "arbol"

# Comparación normal (basada en ASCII/Unicode)
print(f"Normal: '{cadena1}' < '{cadena2}': {cadena1 < cadena2}")

# Comparación sensible al locale
# locale.strcoll devuelve <0 si s1<s2, 0 si s1==s2, >0 si s1>s2
print(f"Locale: '{cadena1}' < '{cadena2}': {locale.strcoll(cadena1, cadena2) < 0}")

# Restablecer el locale predeterminado si es necesario
locale.setlocale(locale.LC_ALL, '')
```
