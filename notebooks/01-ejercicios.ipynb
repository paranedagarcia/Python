```python
"""3. Extracción de patrones con Expresiones Regulares (Regex)
Cuando el análisis requiere buscar formatos específicos (como correos o URLs), se utiliza el módulo re. Esto permite no solo validar si el patrón existe, sino extraer fragmentos específicos mediante "grupos"
"""
# Análisis: Extraer el dominio de una dirección de correo electrónico.
import re
email = "usuario@google.com"
patron = r"[a-z0-9._%+-]+@([a-z0-9.-]+\.[a-z]{2,})"
match = re.match(patron, email, re.IGNORECASE)
if match:
    dominio = match.group(1) # Extrae lo que está en el primer paréntesis [11, 12]
# Resultado: 'google.com'
dominio
```
