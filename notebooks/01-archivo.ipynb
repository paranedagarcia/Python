```python
import pandas as pd
 # mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
# leer archivo excel
filepath = "../data/PLAN DE COMPRAS 2025.xlsx"
df = pd.read_excel(filepath)
df.head()
```

```python
# total de 'monto total Ítem Año 2025' por 'nombre responsable'
total_por_responsable = df.groupby('Nombre responsable')['Monto Total Ítem Año 2025'].sum().reset_index()
total_por_responsable = total_por_responsable.sort_values(by='Monto Total Ítem Año 2025', ascending=False)
total_por_responsable.head()
```

```python
# total de 'monto total Ítem Año 2025' por 'nombre responsable' y 'tipo de gasto'
total_por_responsable_tipo = df.groupby(['Nombre responsable', 'Tipo de Gasto'])['Monto total Ítem Año 2025'].sum()
```
